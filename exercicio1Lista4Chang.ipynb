```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 
import requests
import datetime
import matplotlib.pyplot as plt
# import plotly.graph_objects as go
import plotly.offline as py
import plotly
from plotly.offline import init_notebook_mode
from plotly import graph_objs as go
plotly.offline.init_notebook_mode(connected=True)
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
# wavefun(level=5)  o parmetro para o nvel de refinamento baseado nos "filter coefficients"
# referncia: https://pywavelets.readthedocs.io/en/latest/regression/wavelet.html#and-now-the-wavefun
# escolhi 5 para uma boa visualizao do grfico

[phi, psi, x] = pywt.Wavelet('db4').wavefun(level=5)

data = [go.Scatter(x=x, y=psi, mode='lines', name="Daubechies 4 psi")]
data += [go.Scatter(x=x, y=phi, mode='lines', name="Daubechies 4 phi")]

[phi, psi, x] = pywt.Wavelet('db8').wavefun(level=4)

data += [go.Scatter(x=x, y=psi, mode='lines', name="Daubechies 8 psi")]
data += [go.Scatter(x=x, y=phi, mode='lines', name="Daubechies 8 phi")]

layout = go.Layout(
    title='Wavelets',
    xaxis={'title':'x'},
    yaxis={'title':'y'}
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
